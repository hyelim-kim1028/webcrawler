```python
# final_version 
# 반복해서 같은것만 출력 

#에러 1 
# 첫번째 페이지가 5번 반복되고, 
# 페이지 6으로 넘어간다 > 그리고 페이지 6이 5번 반복된다 
# 하나씩 눌리고 있지 않다 

#         button = driver.find_element_by_link_text(f'{str(page_num)}')
#         button.click()
# 이유: button.click() 없어서 눌리지 않았다 

# 에러 2 
# 페이지는 잘 찍히는데 그게 다 데이터베이스가 되지 않는다 

while_code = 0
page_num = 1

while True:
    if while_code == 1:
        break
#         target = 'https://korean.visitkorea.or.kr/search/search_list.do?keyword=%ED%9E%90%EB%A7%8
    html = driver.page_source #get the source of the current page # HTML 긁어오기
    soup = BeautifulSoup(html, 'html.parser')
    item_box = soup.find(class_='list_thumType type1').find_all(name = 'li', attrs = {'class': False})
   
    for item in item_box:
        try:
            tit = item.find(attrs = {'class':'tit'}).text
            loc = item.find(attrs = {'class': 'service'}).text
            tag = item.find(attrs = {'class':'tag_type'}).text

            print(tit)
            print(loc)
            print(tag)

        except:
            if page_num > 10:
                while_code = 1
                break
            
            pass
        
        titList.append(tit)
        locList.append(loc)
        tagList.append(tag)
    
    if page_num % 5 != 0: 
        page_num += 1
        print(f"page_num5: {page_num}")
        button = driver.find_element_by_link_text(f'{str(page_num)}')
        button.click()
        time.sleep(1)
        print("scout2")
        
    elif page_num % 5 == 0:
        page_num += 1
        print(f"page_num0: {page_num}")
        driver.find_element_by_link_text("다음").click()
        time.sleep(1)
        print("scout3")
        
    if page_num > 10 :
        while_code = 1
        break
```
